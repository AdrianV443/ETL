```python
def get_data(i):
    clase_act_id = i["CLASE_ACTIVIDAD_ID"]
    empresa = i['Nombre']
    clase_act = i['Clase_actividad']
    estrato = i["Estrato"]
    calle = i['Calle']
    numero_ext = i['Num_Exterior']
    colonia = i['Colonia']
    cp = i['CP']
    ubicacion = i['Ubicacion']
    longitud = i['Longitud']
    latitud = i['Latitud']
    fecha_alta = i["Fecha_Alta"]
    telefono = i['Telefono']
    correo = i["Correo_e"]
    return clase_act_id,empresa,clase_act,estrato,calle,numero_ext,colonia,cp,ubicacion,longitud,latitud,fecha_alta,telefono,correo
```

```python
# df
base1 = []
base2 = []
base3 = []
# base4 = []
# base5 = []
# base6 = []
col = ['clase_act_id','empresa','clase_act','estrato','calle','numero_ext','colonia','cp','ubicacion','longitud','latitud','fecha_alta','telefono','correo']
for i in data1:
    base1.append(get_data(i))
df1 = pd.DataFrame(base1, columns=col)
for i in data2:
    base2.append(get_data(i))
df2 = pd.DataFrame(base2, columns=col)
for i in data3:
    base3.append(get_data(i))
df3 = pd.DataFrame(base3, columns=col)
# for i in data4:
#     base4.append(get_data(i))
# df4 = pd.DataFrame(base4, columns=col)
# for i in data5:
#     base5.append(get_data(i))
# df5 = pd.DataFrame(base5, columns=col)
# for i in data6:
#     base6.append(get_data(i))
# df6 = pd.DataFrame(base6, columns=col)
```
